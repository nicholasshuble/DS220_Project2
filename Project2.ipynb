```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("collegefootballbowl.csv")
```

```python
data.head()
```

```python
PSUwin = data[data['winner_tie'].str.contains('Penn State')]
print(PSUwin)
print(len(PSUwin))
```

```python
PSUloss = data[data['loser_tie'].str.contains('Penn State')]
print(PSUloss)
print(len(PSUloss))
```

```python
PSUwtie = ((PSUwin['winner_points'] == PSUwin['loser_points']))
print(PSUwtie)
count = 0
for bool in PSUwtie:
    if bool == True:
        count += 1
print(count)
```

```python
PSUltie = ((PSUloss['winner_points'] == PSUloss['loser_points']))
print(PSUltie)
count = 0
for bool in PSUltie:
    if bool == True:
        count += 1
print(count)
```

**Penn State's Bowl Record**

Penn State has won or tied 32 games. Of those 32 games, two of them resulted in both teams having the same score, meaning they were a actually a tie and not a win for Penn State. So Penn State has loss 19 bowl games. We can then conclude that Penn State's total bowl game record as Wins-Losses-Ties is 30-19-2.
